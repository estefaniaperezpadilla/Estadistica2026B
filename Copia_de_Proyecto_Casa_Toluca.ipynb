```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.multicomp import pairwise_tukeyhsd

link = "https://raw.githubusercontent.com/estefaniaperezpadilla/Estadistica2026B/refs/heads/main/Casa_toluca1/Casa_toluca.csv"
df = pd.read_csv(link)

# limpiar columnas
df.columns = df.columns.str.strip()

# Filtrar ubicaciones
top_ubicaciones = df['Ubicación'].value_counts().head(6).index
df_filtrado = df[df['Ubicación'].isin(top_ubicaciones)].copy()

# Guardar nombres originales
df_filtrado["Ubicación_nombre"] = df_filtrado["Ubicación"]

# Convertir a números
df_filtrado["Ubicación"] = df_filtrado["Ubicación"].astype("category").cat.codes

# Crear diccionario número -> nombre
mapa = dict(enumerate(df_filtrado["Ubicación_nombre"].astype("category").cat.categories))

# Tukey
tukey = pairwise_tukeyhsd(
    endog=df_filtrado['Precio'],
    groups=df_filtrado['Ubicación'],
    alpha=0.05
)

# Extraer datos
data = tukey._results_table.data[1:]

# 👇 usar nombres en vez de números
grupos = [f"{mapa[row[0]]} - {mapa[row[1]]}" for row in data]

meandiff = [row[2] for row in data]
lower = [row[4] for row in data]
upper = [row[5] for row in data]

y_pos = np.arange(len(grupos)) * 1.5

# Gráfico
plt.figure(figsize=(10,12))

for i in range(len(y_pos)):
    plt.plot([lower[i], upper[i]], [y_pos[i], y_pos[i]], 'k-')

plt.plot(meandiff, y_pos, 'ko')

plt.yticks(y_pos, grupos, fontsize=8)
plt.xlabel('Diferencia de medias')
plt.title('Tukey - Precio vs Ubicación')

plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import r2_score
import pandas as pd
import statsmodels.api as sm

# Cargar datos
link = "https://raw.githubusercontent.com/estefaniaperezpadilla/Estadistica2026B/refs/heads/main/Casa_toluca1/Casa_toluca.csv"
df = pd.read_csv(link)

# Renombrar columnas para evitar errores
df.columns = df.columns.str.replace("ñ","n")

# Variables
x = df[["Area","Banos","Habitaciones"]]
y = df["Precio"]

# Convertir a numérico
x = x.apply(pd.to_numeric, errors='coerce')
y = pd.to_numeric(y, errors='coerce')

# Eliminar valores nulos
datos = pd.concat([x,y], axis=1).dropna()
x = datos[["Area","Banos","Habitaciones"]]
y = datos["Precio"]

# Modelo
x_constante = sm.add_constant(x)
modelo = sm.OLS(y, x_constante).fit()
y_calculada = modelo.predict(x_constante)

# Convertir a array (esto evita el error)
r2 = r2_score(y.values, y_calculada.values)

print(f'Coeficiente de determinación: {r2:0.2%}')
```

```python
# 7. Obtenga un intervalo de confianza del 95% para la pendiente de la recta de
# regresión ajustada ( b1 )

# intervalo de confianza = 1 - alpha

modelo.conf_int(alpha = 0.05)
```

```python
# 8. Calcule los residuales y trace un nuevo gráfico de dispersión. Comente,
# ¿Parece que se verifican los supuestos?

# Los residuales son la diferencia entre el valor real y el valor que da el
# modelo

# residuales = y - y_calculada
residuales = modelo.resid

residuales
```

```python
#shapiro
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns

# Prueba de Shapiro-Wilk
stat, valor_p_sh = shapiro(residuales)
print(f"valor-p (Shapiro) = {valor_p_sh}")

# Visualización: Q-Q plot
sm.qqplot(residuales, line='s')
plt.title("Q-Q plot de residuos")
plt.show()

# Histograma
sns.histplot(residuales, kde=True)
plt.title("Histograma de residuos")
plt.show()
```

```python
# 10. Prueba de Breusch-Pagan
# H0: No hay heterocedasticidad
# H1: Hay heterocedasticidad
# alpha = 0.05

import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.api import het_breuschpagan

# Cargar datos
link = "https://raw.githubusercontent.com/estefaniaperezpadilla/Estadistica2026B/refs/heads/main/Casa_toluca1/Casa_toluca.csv"
df = pd.read_csv(link)

# Renombrar columnas para evitar errores
df.columns = df.columns.str.replace("ñ","n")

# Variables
x = df[["Area","Banos","Habitaciones"]]
y = df["Precio"]

# Convertir a numérico
x = x.apply(pd.to_numeric, errors='coerce')
y = pd.to_numeric(y, errors='coerce')

# Eliminar valores nulos
datos = pd.concat([x, y], axis=1).dropna()
x = datos[["Area","Banos","Habitaciones"]]
y = datos["Precio"]

# Modelo
x_constante = sm.add_constant(x)
modelo = sm.OLS(y, x_constante).fit()

# Residuales
residuales = modelo.resid

# Test Breusch-Pagan
bp_test = het_breuschpagan(residuales, x_constante)
valor_p_bp = bp_test[1]

print(f'valor_p de Breusch-Pagan: {valor_p_bp:0.4f}')

# Interpretación automática
if valor_p_bp > 0.05:
    print("No hay heterocedasticidad")
else:
    print("Hay heterocedasticidad")
```

```python
# 12. Tabla ANOVA
# H0: No hay relación entre variables
# H1: Sí hay relación

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Cargar datos
link = "https://raw.githubusercontent.com/estefaniaperezpadilla/Estadistica2026B/refs/heads/main/Casa_toluca1/Casa_toluca.csv"
df = pd.read_csv(link)

# Renombrar columnas para evitar error
df.columns = df.columns.str.replace("ñ","n")

# Convertir a numérico
df["Area"] = pd.to_numeric(df["Area"], errors="coerce")
df["Banos"] = pd.to_numeric(df["Banos"], errors="coerce")
df["Habitaciones"] = pd.to_numeric(df["Habitaciones"], errors="coerce")
df["Precio"] = pd.to_numeric(df["Precio"], errors="coerce")

# Eliminar NA
df = df.dropna()

# Modelo ANOVA
modelo_lineal = ols('Precio ~ Area + Banos + Habitaciones', data=df).fit()

# Tabla ANOVA
tabla_anova = sm.stats.anova_lm(modelo_lineal)

print(tabla_anova)
```
